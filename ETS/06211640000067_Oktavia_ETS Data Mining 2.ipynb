```python
# Import Libraries
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# Import Dataset
retail = pd.read_csv('online_retail.csv')
retail.tail(n=5)
```

```python
retail['Description'] = retail['Description'].str.strip()
retail.dropna(axis=0, subset=['InvoiceNo'], inplace=True)

retail['InvoiceNo'] = retail['InvoiceNo'].astype('str')
retail = retail[~retail['InvoiceNo'].str.contains('C')]
```

```python
basket = (retail.groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
basket.tail(n=5)
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
basket_sets = basket.applymap(encode_units)
 
# Delete Postage Column
basket_sets.drop('POSTAGE', inplace=True, axis=1)
basket_sets.head()
```

```python
# Build up the Frequent Items (Min. Support 0.03)
frequent_itemsets = apriori(basket_sets, min_support=0.03, use_colnames=True)
frequent_itemsets
```

```python
# Create the Rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
# Filter Lift (6) and Confidence (0.6)
rules[ (rules['lift'] >= 6) &
       (rules['confidence'] >= 0.6) ]
```
