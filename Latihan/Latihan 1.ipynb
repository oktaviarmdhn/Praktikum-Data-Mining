## Data Integration

```python
import pandas as pd
import numpy as np
```

```python
# Merging
mhs = pd.DataFrame({'ID' : ['001','002','003','004','005'],
                    'JK' : ['P','P','L','P','L'],
                    'Usia' : [17,18,16,19,17]})
nilai = pd.DataFrame({'ID' : ['001','002','003','004','005'],
                      'IPK' : ['3.94','4.00','3.22','3.38','3.00'],
                      'Nilai' : ['A','A','AB','AB','B']})
dat = pd.merge(mhs,nilai,on='ID')
dat
```

```python
# Merging Different Key ID
mhs = pd.DataFrame({'ID' : ['001','002','003','004','005'],
                    'JK' : ['P','P','L','P','L'],
                    'Usia' : [17,18,16,19,17]})
nilai = pd.DataFrame({'ID' : ['002','001','005','003'],
                      'IPK' : ['3.94','4.00','3.22','3.38'],
                      'Nilai' : ['A','A','AB','AB']})
data = pd.merge(mhs,nilai,on='ID')
data
```

```python
# Ordering Variable
data = data[['ID','Usia','JK','IPK','Nilai']]
data
```

```python
# Multiple Key
mhs2 = pd.DataFrame({'ID': ['001', '001', '002', '002'],
                     'Kelas': ['Datmin', 'Multivariat', 'Datmin', 'ADW'],
                     'JK': ['P', 'P', 'L', 'L'], 
                     'Usia': ['19', '21', '17', '23']})

nilai2 = pd.DataFrame({'ID': ['001', '001', '002', '002'],
                      'Kelas': ['ADK', 'Pengayaan', 'Datmin', 'ADW'],
                      'IPK': ['3.94', '3.34', '3.75', '3.11'],
                      'Nilai': ['A', 'AB', 'AB', 'B']})

data2 = pd.merge(mhs2, nilai2, on=['ID', 'Kelas'])
data2
```

```python
# Merge methods (default how = "left")
data_left = pd.merge(mhs2, nilai2, how='left', on=['ID', 'Kelas'])
data_left
```

```python
# Merge methods (default how = "right")
data_right = pd.merge(mhs2, nilai2, how='right', on=['ID', 'Kelas'])
data_right
```

```python
# Merge methods (default how = "outer")
data_outer = pd.merge(mhs2, nilai2, how='outer', on=['ID', 'Kelas'])
data_outer
```

```python
# Merge methods (default how = "inner")
data_inner = pd.merge(mhs2, nilai2, how='inner', on=['ID', 'Kelas'])
data_inner
```

```python
# Joining
mhs = pd.DataFrame({'JK': ['P', 'P', 'L'], 
                      'Usia': ['19', '20', '20']},
                    index=['001', '002', '003'])
nilai = pd.DataFrame({'Kelas': ['ADW', 'ADK', 'Datmin'],
                      'IPK': ['4.00', '3.44', '3.50']},
                    index=['001', '002', '003'])
data_j = mhs.join(nilai)
data_j
```

```python
# Concatenating
df1 = pd.DataFrame({'A': ['A1', 'A2', 'A3', 'A4'],
                'B': ['B1', 'B2', 'B3', 'B4'],
                'C': ['C1', 'C2', 'C3', 'C4'],
                'D': ['D1', 'D2', 'D3', 'D4']})
df2 = pd.DataFrame({'A': ['A5', 'A6', 'A7', 'A8'],
                'B': ['B5', 'B6', 'B7', 'B8'],
                'C': ['C5', 'C6', 'C7', 'C8'],
                'D': ['D5', 'D6', 'D7', 'D8']})
df3 = pd.DataFrame({'A': ['A9', 'A10', 'A11', 'A12'],
                'B': ['B9', 'B10', 'B11', 'B12'],
                'C': ['C9', 'C10', 'C11', 'C13'],
                'D': ['D9', 'D10', 'D11', 'D12']})
frames = [df1, df2, df3]
data3 = pd.concat(frames)
data3
```

```python
# Add Category
data3 = pd.concat(frames, keys = ['1','2','3'])
data3
```

```python
# Default join=outer (Join All Rows)
df4 = pd.DataFrame({'B': ['B3', 'B4', 'B5', 'B6'],
                    'D': ['D3', 'D4', 'D5', 'D6'],
                    'F': ['F3', 'F4', 'F5', 'F6']},
                    index=[3, 4, 5, 6])

data4 = pd.concat([df1, df4], axis=0,sort=False)
data4
```

```python
# Concatenating using Append
data5 = df1.append(df2)
data5
```

```python
data6 = df1.append(df4)
data6
```

```python
data7 = df1.append([df2,df3])
data7
```
