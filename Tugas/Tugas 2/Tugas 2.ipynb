## Imputation Missing Value

```python
# Count Missing Value of Each Variable
np.sum(data.isnull())
```

```python
data.isnull().sum().sum()
```

```python
# Missing Value Imputation of Numeric Data
data['Medicaid Enrollment (2013)'] = data['Medicaid Enrollment (2013)'].fillna((data['Medicaid Enrollment (2013)'].mean())) 
data['Medicaid Enrollment Change (2013-2016)'] = data['Medicaid Enrollment Change (2013-2016)'].fillna((data['Medicaid Enrollment Change (2013-2016)'].mean()))
data.isnull().any()
```

```python
# Imputation for Categorical Data with Mode (based on each column)
data = data.apply(lambda x:x.fillna(x.value_counts().index[0]))
data.isnull().any()
```

## Encode Labels

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Encode without Create a New Column
data['State Medicaid Expansion (2016)'] = LabelEncoder().fit_transform(data['State Medicaid Expansion (2016)'])
data.head(n=5)
```

## Standardize

```python
from sklearn.preprocessing import StandardScaler
```

```python
data['Medicaid Enrollment (2013)'] = StandardScaler().fit(data[['Medicaid Enrollment (2013)']]).transform(data[['Medicaid Enrollment (2013)']])
data['Medicaid Enrollment (2016)'] = StandardScaler().fit(data[['Medicaid Enrollment (2016)']]).transform(data[['Medicaid Enrollment (2016)']])
data['Medicaid Enrollment Change (2013-2016)'] = StandardScaler().fit(data[['Medicaid Enrollment Change (2013-2016)']]).transform(data[['Medicaid Enrollment Change (2013-2016)']])
data.head(n=5)
```

## Normalize

```python
from sklearn.preprocessing import Normalizer 
```

```python
data['Medicaid Enrollment (2013)'] = Normalizer().fit(data[['Medicaid Enrollment (2013)']]).transform(data[['Medicaid Enrollment (2013)']])
data['Medicaid Enrollment (2016)'] = Normalizer().fit(data[['Medicaid Enrollment (2016)']]).transform(data[['Medicaid Enrollment (2016)']])
data['Medicaid Enrollment Change (2013-2016)'] = Normalizer().fit(data[['Medicaid Enrollment Change (2013-2016)']]).transform(data[['Medicaid Enrollment Change (2013-2016)']])
data.head(n=5)
```
