```python
import pandas as pd
import numpy as np
```

```python
# Load Data
births1 = pd.read_csv(r'births1.csv')
births2 = pd.read_csv(r'births2.csv')
births1
```

```python
births2
```

```python
# Concatenate Data
frames = [births1,births2]
births = pd.concat(frames)
births
```

```python
# Add Category
birthscat = pd.concat(frames, keys=['Statistical Area 2', 'Area Unit'])
birthscat
```

```python
# Insert New Column
births['Total'] = (births['2013']+births['2014']+births['2015']+births['2016']+births['2017'])
births
```

```python
# Delete Column
births = births.drop(['code'], axis=1)
births
```

```python
# Show the Geography Type and Total Only / Filter Column
geototal = births.drop(['description','2013','2014','2015','2016','2017'], axis=1)
geototal
```
